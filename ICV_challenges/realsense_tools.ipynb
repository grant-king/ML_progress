```python
import pyrealsense2 as rs
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
points = rs.points()
pipeline = rs.pipeline()
config = rs.config()
config.enable_stream(rs.stream.infrared, 1)
config.enable_stream(rs.stream.infrared, 2)
```

```python
profile = pipeline.start(config)
running = True

while running:
    frames = pipeline.wait_for_frames()
    left_frame = frames.get_infrared_frame(1)
    right_frame = frames.get_infrared_frame(2)
    if left_frame and right_frame:
        running = False
pipeline.stop()
```

```python
left_image = np.asanyarray(left_frame.get_data())
right_image = np.asanyarray(right_frame.get_data())

cv2.imwrite('left2.png', left_image)
cv2.imwrite('right2.png', right_image)

fig, (sp1, sp2) = plt.subplots(1, 2, figsize=[15, 10])
sp1.imshow(left_image, cmap='gray')
sp2.imshow(right_image, cmap='gray')

_ = [ax.set_axis_off() for ax in [sp1, sp2]]
```
